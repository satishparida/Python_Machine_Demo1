
Masked wordcloud
================

Using a mask you can generate wordclouds in arbitrary shapes.



```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS #https://pypi.python.org/pypi/wordcloud #https://pypi.python.org/pypi/stopwords/0.1.3

d = path.dirname("./")

# Read the whole text.
#text = open(path.join(d, 'alice.txt')).read()
text = open(path.join(d, 'accentureTechvision2018.txt')).read()
#accentureTechvision2018.txt
```

- Read Stopwords
- Store mask as numpy array

```python
# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open(path.join(d, "accenture.png")))


stopwords = set(STOPWORDS)
stopwords.add("said") # Add another common word or custom word
```

Use WordCloud package and specify background color, max words to be displayed, mask to be used and stop words that need to be removed from the text
- 

```python
wc = WordCloud(background_color="green", max_words=200, mask=alice_mask,
               stopwords=stopwords) #How many words, Which mask to use
# generate word cloud
wc.generate(text)

# store to file
#wc.to_file(path.join(d, "alice.png"))
wc.to_file(path.join(d, "accenture-1.png"))
```

```python
# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
#plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
#plt.axis("off")
```
