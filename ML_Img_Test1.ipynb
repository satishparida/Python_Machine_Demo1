```python
#required to plt.show()
%matplotlib inline
import matplotlib.image as mpimg
import numpy as np
import math

img = mpimg.imread('stock-photo.jpg')

from matplotlib import pyplot as plt
plt.imshow(img)
plt.show()

D1_inten_per_list=np.true_divide(img,2.55)

D1_mean_list_np=img.mean(2)
D1_mean_list_diff_np = np.zeros(shape=(D1_mean_list_np[:,0].shape[0],D1_mean_list_np[0,:].shape[0],2))

for i in range(1,D1_mean_list_np[:,0].shape[0]-2):
    D1_mean_list_diff_np[i,:,0]=D1_mean_list_np[i,:]-D1_mean_list_np[i-1,:]
for i in range(1,D1_mean_list_np[0,:].shape[0]-2):
    D1_mean_list_diff_np[:,i,1]=D1_mean_list_np[:,i]-D1_mean_list_np[:,i-1]

D1_mean_list_diff_mean_np=D1_mean_list_diff_np.mean(2)

plt.imshow(D1_mean_list_diff_mean_np)
plt.show()

D1_mean_list_diff_mean_final_np = np.zeros(shape=(D1_mean_list_np[:,0].shape[0],D1_mean_list_np[0,:].shape[0],3))

for i in range(1,D1_mean_list_diff_mean_final_np[:,0].shape[0]-2):
    D1_mean_list_diff_mean_final_np[i,:,0]=D1_mean_list_diff_mean_np[i,:].astype(int)

D1_mean_list_diff_mean_np
```
